```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
weather_data = pd.read_csv('weather - weather.csv')
weather_data
```

```python
weather_data.info()
```

```python
weather_data.isnull().sum()
```

```python
weather_data['WindDir9am'] = weather_data['WindDir9am'].fillna(method='ffill')
weather_data['WindDir3pm'] = weather_data['WindDir3pm'].fillna(method='ffill')
weather_data['Sunshine'] = weather_data['Sunshine'].fillna(method='ffill')
weather_data['WindGustSpeed'] = weather_data['WindGustSpeed'].fillna(method='ffill')
weather_data['WindGustDir'] = weather_data['WindGustDir'].fillna(method='ffill')
weather_data['WindSpeed9am'] = weather_data['WindSpeed9am'].fillna(method='ffill')
```

```python
weather_data[['Sunshine','WindDir9am','WindDir3pm','WindGustDir','WindSpeed9am','WindGustSpeed']].isnull().sum()
```

```python
weather_data.shape
```

```python
weather_data.size
```

```python
weather_data.describe()
```

```python
stats = weather_data[['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation']].agg(['mean', 'median', 'std'])
stats
```

```python
# Plotting the weather variables over time (using row index)
plt.figure(figsize=(14, 7))  # Set the figure size
plt.plot(weather_data.index, weather_data['MinTemp'], label='MinTemp')
plt.plot(weather_data.index, weather_data['MaxTemp'], label='MaxTemp')
plt.plot(weather_data.index, weather_data['Rainfall'], label='Rainfall')
plt.plot(weather_data.index, weather_data['Evaporation'], label='Evaporation')
plt.plot(weather_data.index, weather_data['Sunshine'], label='Sunshine')

# Adding labels and title
plt.xlabel('Row Index')
plt.ylabel('Value')
plt.title('Variations in Weather Variables')
plt.legend()

# Display the plot
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

```python
#correlation analysis
# Select the relevant columns
columns = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation']
data = weather_data[columns]

# Calculate the correlation matrix
corr_matrix = data.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')

# Show the plot
plt.show()
```

```python
# Create a histogram of the rainfall data
plt.figure(figsize=(8, 6))
plt.hist(weather_data['Rainfall'], bins=20, edgecolor='black')
plt.xlabel('Rainfall (mm)')
plt.ylabel('Frequency')
plt.title('Distribution of Rainfall')

# Show the plot
plt.show()
```

```python
# Create a kernel density plot of the rainfall data
plt.figure(figsize=(8, 6))
sns.kdeplot(data=weather_data, x='Rainfall', shade=True)
plt.xlabel('Rainfall (mm)')
plt.ylabel('Density')
plt.title('Distribution of Rainfall')

# Show the plot
plt.show()
```

```python
# Calculate the average values of weather variables
avg_values = weather_data.mean()

# Create a bar graph
plt.figure(figsize=(12, 6))
avg_values.plot(kind='bar')
plt.xlabel('Weather Variables')
plt.ylabel('Average Value')
plt.title('Average Values of Weather Variables')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
```
